```python
def prep_list(df):
    Names_L= df['Author_I'].values
    title_L = df['Paper Title'].values
    affiliations_L = df["Presenter's Affiliation"].values
    abstract_L = df['Abstract'].values
    abstract_n_L=df['Abstract Number'].values
    Date_L=df['Date/Time'].values
    Date_L_N=df['Nepal Date/Time'].values
    return Names_L,affiliations_L,title_L,abstract_L,abstract_n_L,Date_L,Date_L_N
```

```python
def create_html(df_name,html_name):
    Names_L,affiliations_L,title_L,abstract_L,abstract_n_L,Date_L,Date_L_N=prep_list(df_name)
    fo = open(html_name, 'w',encoding="utf-8")
    fo.write(prepend)
    fo = open(html_name, 'a+',encoding="utf-8")
    for i in range(len(Names_L)):

        names_ = Names_L[i]
        title_ = title_L[i]
        affil_ = affiliations_L[i]
        abstract_ = abstract_L[i]
        abstract_n_=abstract_n_L[i]
        date_=Date_L[i]
        nepaldate_=Date_L_N[i]

        html_code = '''

        <!-- ########################################################################## -->

        <tr>
                    <td width="158" colspan="2">
                        <p> Date/Time: <br>
                            ET: {date_} </br>
                            Nepal: {nepaldate_}
                        </p>
                    </td>

        <td width="521" colspan="6">
            <p>Abstract Number: {abstract_n_} </p>

            <p>Presenting Author: {names_}
            </p>
            
            <p>Presenter's Affiliation: {affil_}
            </p>

            <p> <strong>Title: </strong>{title_}</p>

            <p class="show_hide">Show/Hide Abstract</p>
            <div class="show_on_click">{abstract_}</div>

        </td>
        </tr>

    '''.format(date_=date_,nepaldate_=nepaldate_,abstract_n_=abstract_n_,affil_=affil_,names_=names_,
               title_ = title_,abstract_=abstract_)
        fo.write(html_code)

    fo = open(html_name, 'a+',encoding="utf-8")
    fo.write(postpend)



    fo.close()

    
    
    
```

```python
create_html(df_data,html_name='html_code_2022_data.html')
create_html(df_PER,html_name='html_code_2022_PER.html')
create_html(df_Condensed,html_name='html_code_2022_Condensed.html')
create_html(df_Bio,html_name='html_code_2022_Bio.html')
create_html(df_Plasma,html_name='html_code_2022_Plasma.html')
create_html(df_HEP,html_name='html_code_2022_HEP.html')
create_html(df_Applied,html_name='html_code_2022_Applied.html')
create_html(df_Astro,html_name='html_code_2022_Astro.html')
create_html(df_Atmos,html_name='html_code_2022_Atmospheric.html')
create_html(df_InPerson,html_name='html_code_2022_InPerson.html')
```

```python
def prep_list(df):
    Names_L= df['Author_I'].values
    title_L = df['Paper Title'].values
    Date_L=df['Start_DateTime'].values
    EndDate_L=df['End_DateTime'].values
    return Names_L,title_L,Date_L,EndDate_L
```

```python
c=Calendar()
Names_L,title_L,Date_L,EndDate_L=prep_list(df)
str(Date_L[0])
for i in range(len(Names_L)):
    e=Event()
    e.name=Names_L[i]
    e.begin=str(Date_L[i])
    e.end=str(EndDate_L[i])
    e.description=title_L[i]
    c.events.add(e)
c.events
```
